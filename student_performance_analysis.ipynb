# Student Performance Data Analysis

This project analyzes student academic performance using Python and Pandas.

## Objectives
- Analyze student marks and attendance
- Identify top performers
- Evaluate performance trends
- Visualize relationships between attendance and marks

## Tools Used
- Python
- Pandas
- Matplotlib




```python
import pandas as pd
```

```python
data = {
    "name": ["Aarav", "Neha","Riya","Kunal", "Mehal"],
    "marks" : [85, 72, 90, 65, 78],
    "attendance": [92, 88, 95, 70, 85]
}
df = pd.DataFrame(data)
df
```

```python
df["marks"].mean()
```

```python
df[df["marks"]== df["marks"].max()]
```

```python
df[df["marks"] > df["marks"].mean()]
```

```python
df.sort_values(by="marks", ascending=False)
```

```python
df[["marks", "attendance"]].mean()
```

```python
import pandas as pd

df = pd.read_csv("students.csv")
df
```

# Insights

- Average marks of students is around 78.
- Students with higher attendence generally score higher score.
- Riya is the top performing student.
- Some student with low attendence show lower performance.

```python
import matplotlib.pyplot as plt
plt.bar(df["name"], df["marks"])
plt.title("Student marks Comparison")
plt.xlabel("Student")
plt.ylabel("Marks")
plt.show()
```

```python
plt.scatter(df["attendence"], df["marks"])

plt.title("Attendence vs Marks")
plt.xlabel("Attendence")
plt.ylabel("Marks")

plt.show()
```

## Visualization Insight

The scatter plot suggests a positive relationship between attendance and marks.
Students with higher attendance tend to achieve better academic performance.
